# 期权

```python
def timestamp2time(now):
    now02 = time.strftime('%Y-%m-%d %H:%M:%S',time.localtime(now/1000))
    return(now02)
```

```python
%%time
#期权数据处理，分理处规范的快照数据和逐笔成交数据
# type = 2卖出
    
def cleaningData():
    # 数据存储在这个文件夹下
    read_path = './jex'
    
    # depth是逐笔成交的数据，market是快照的数据，然后其他的数据暂时没有搞定，就先不使用了
    depth,market = {}, {}
    count0,count1 = 0,0
    labels = ['datetime','open','high','low','close','vol']
    
    # 获取文件夹内的列表
    for filename in os.listdir(read_path):
        # 清洗文件
        with open(read_path + r'/' + filename, 'r') as f:
            data = f.readlines()
            
        date = filename.split('_')[0]
        date = date[:4] + '-' + date[4:6] + '-'+ date[6:]

        for i in data:
            lineData = eval(i[2:])
            if lineData[0][:5] == 'depth':
                if 'deal' in lineData[1].keys():
                    for j in range(len(lineData[1]['deal'])):
                        a = lineData[1]['deal'][j]
                        a['date'] = date + ' ' + a['date']
                        a['kind'] = lineData[0]
                        depth[count0] = a
                        count0+=1

                    
            elif lineData[0][:6] == 'market':
                a = {}
                for j in range(len(lineData[1])):
                    a['kind'] = lineData[0]
                    for i in range(6):
                        if i == 0:
                            a[labels[i]] = timestamp2time(lineData[1][j][i])
                        else:
                            a[labels[i]] = lineData[1][j][i]
                    market[count1] = a
                    count1 += 1
                
        print(filename,' done.')
    return depth, market
```

```python
%%time
depth, market = cleaningData()

market = pd.DataFrame(market).T
depth = pd.DataFrame(depth).T

market.to_csv('market.csv',index = 0,header = None)
depth.to_csv('depth.csv',index = 0,header = None)
```

# 现货——快照数据

```python
def cleaningHuobi():
    def cleaningOne(filename):
        path = r'./huobi/'
        digitCurrency = {}
        count = 0
        labels = ['datetime','kinds','open','high','low','close','vol']
        with open(path + filename, 'r') as f:
            data = f.readlines()
        for i in data:
            i = i.split(',')
            a = {}
            for j in range(7):
                a[labels[j]] = i[j].strip()
            digitCurrency[count] = a
            count+=1
        
        spot = pd.DataFrame(digitCurrency).T
        spot.to_csv('./csv/{}_spot.csv'.format(filename))
    
    path = r'./huobi/'
    Parallel(8)(delayed(cleaningOne)(filename) for filename in os.listdir(path))
```

```python
%%time
cleaningHuobi()
```

```python
ans = []
for i in os.listdir('./csv')[1:]:
    if i[-9:] == '_spot.csv':
        a = pd.read_csv('./csv/{}'.format(i),index_col = 0,header = 0)
        ans.append(a)
ans = pd.concat(ans)
```

```python
ans.to_csv('./csv/spot.csv',index = None,header = None)
```
